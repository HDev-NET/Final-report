```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
path_data = 'data/healthcare-dataset-stroke-data.csv'

df = pd.read_csv(path_data)
df.head()
```

```python
df.shape
```

```python
# Data information
df.info()
```

```python
# Missing data
df.isnull().sum()
```

## Exploratory Analysis

```python
fig = px.histogram(df, x="bmi", color="stroke", marginal="box",
                         hover_data=df.columns, color_discrete_map={1: '#ea4335', 0: '#4285f4'})
fig.show()
print('Những người bị đột quỵ có bmi cao hơn một chút so với những người chưa từng bị đột quỵ')
```

```python
fig = px.histogram(df, x="age", color="stroke", marginal="box",
                         hover_data=df.columns, color_discrete_map={1: '#ea4335', 0: '#4285f4'})
fig.show()

print("Hầu hết những người đột quỵ là người già")
```

```python
fig = px.histogram(df, x="avg_glucose_level", color="stroke", marginal="box",
                         hover_data=df.columns, color_discrete_map={1: '#ea4335', 0: '#4285f4'})
fig.show()

print('Hầu hết những người bị đột quỵ thường có mức đường huyết trung bình cao hơn những người chưa bao giờ bị đột quỵ')
```
